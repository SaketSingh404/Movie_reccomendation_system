```python
def list_countries(obj):
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
def list_keywords(obj):
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
def list_cast(obj):
    L=[]
    for i in obj[:4]:
        L.append(i['name'])
    return L
```
