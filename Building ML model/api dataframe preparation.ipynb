## Fixing path

```python
def fix_path(path):
    if pd.isna(path):
        path = 'https://picsum.photos/seed/picsum/200/300'
    else:
        path = 'https://image.tmdb.org/t/p/original' + path 
    return path 
    
```

```python
'https://image.tmdb.org/t/p/original'+ new_df["poster_path"][0]
```

```python
new_df.loc[:,"poster_path"] = new_df["poster_path"].apply(fix_path)
```

## Testing Model

```python
import pickle
```

```python
with open('similarity.pkl', 'rb') as f:
    similarity_matrix = pickle.load(f)
```

```python
def recommend(index):
    distances = similarity_matrix[index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:11]
    recommended_index = [i[0] for i in movies_list]
    return new_df.iloc[recommended_index]
```
