```python
def convert(obj,genre=genres):
    l = []
    for i in ast.literal_eval(obj):
        for j in range(len(genre)):
            if i == genre[j]['id']:
                l.append(genre[j]['name'])
    return l
```

```python
df.loc[:,'genre_ids']= df['genre_ids'].apply(convert)
```

```python
def list_countries(obj):
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
def list_keywords(obj):
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
def list_cast(obj):
    L=[]
    for i in obj[:4]:
        L.append(i['name'])
    return L
# def additional_info(movie_id):
#     url = f"https://api.themoviedb.org/3/movie/{movie_id}"
#     params = {"append_to_response": "credits,keywords"}
#     response = requests.get(url, headers=headers, params=params)
#     data = response.json()
#     production_countries = list_countries(data['production_countries'])
#     cast = list_cast(data['credits']['cast'])
#     keywords = list_keywords(data['keywords']['keywords'])
#     return production_countries, cast, keywords
# def additional_info(movie_id):
#     url = f"https://api.themoviedb.org/3/movie/{movie_id}"
#     params = {"append_to_response": "credits,keywords"}

#     response = requests.get(url, headers=headers, params=params)

#     if response.status_code != 200:
#         return [], [], []

#     data = response.json()

#     production_countries = list_countries(data.get('production_countries', []))
#     cast = list_cast(data.get('credits', {}).get('cast', []))
#     keywords = list_keywords(data.get('keywords', {}).get('keywords', []))

#     return production_countries, cast, keywords
def additional_info(movie_id):
    try:
        url = f"https://api.themoviedb.org/3/movie/{movie_id}"
        params = {"append_to_response": "credits,keywords"}
        response = requests.get(url, headers=headers, params=params, timeout=10)
        if response.status_code != 200:
            return [], [], []
        data = response.json()
        production_countries = list_countries(data.get('production_countries', []))
        cast = list_cast(data.get('credits', {}).get('cast', []))
        keywords = list_keywords(data.get('keywords', {}).get('keywords', []))
        return production_countries, cast, keywords
    except:
        return [], [], []
```

# Processing Completed Data

```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv('all_features.csv')
```

```python
df = df.drop(['original_language'],axis=1)
```

```python
df
```

```python
df['production_countries'] = df['production_countries'].apply(lambda x : [i.replace(" ","") for i in ast.literal_eval(x)])
df['cast'] = df['cast'].apply(lambda x : [i.replace(" ","") for i in ast.literal_eval(x)])
df['keywords'] = df['keywords'].apply(lambda x : [i.replace(" ","") for i in ast.literal_eval(x)])
```

```python
df['genre_ids'] = df['genre_ids'].apply(ast.literal_eval)
df['overview'] = df['overview'].apply(ast.literal_eval)
```

```python
df['tags'] = df['overview'] +df['genre_ids'] + df['production_countries'] + df['cast'] + df['keywords']
```

```python
new_df = df[['id','title','tags']]
```
