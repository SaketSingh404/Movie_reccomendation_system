```python
import pandas as pd 
import requests
```

```python
movie_df = pd.DataFrame(columns=['adult',	'backdrop_path',	'genre_ids',	'id',	'original_language',	'original_title',	'overview',	'popularity',	'poster_path',	'release_date',	'title',	'video',	'vote_average',	'vote_count'])
```

```python
movie_df
```

```python
for i in range(1,501):
    url = f"https://api.themoviedb.org/3/discover/movie?include_adult=false&include_video=false&page={i}&sort_by=popularity.desc"

    headers = {
        "accept": "application/json",
        "Authorization": "YOUR_TOKEN"
    }
    print(f'Iteration:{i}')
    response = requests.get(url, headers=headers)
    df = pd.DataFrame(response.json()['results'])
    movie_df = pd.concat([movie_df,df],axis=0,ignore_index=True)
```

```python
movie_df
```

```python
movie_df.to_csv("movies_dataset.csv",index=False)
```
